```python
import pandas as pd
import numpy as np

# ==========================================================
# REPORTE TÉCNICO: COMPLETITUD Y CORRECTITUD (FASE 1)
# ==========================================================

def reporte_calidad_datos(df):
    print("=== ESTUDIO DE CALIDAD DE DATOS (DATA QUALITY ASSESSMENT) ===\n")

    # 1. EVALUACIÓN DE COMPLETITUD (Nulos Técnicos vs Semánticos)
    nulos_tecnicos = df.isnull().sum().sum()
    nulos_semanticos = (df == "-").sum().sum() # Los guiones que detectamos antes
    total_celdas = df.size

    print(f"--- 1. COMPLETITUD ---")
    print(f"- Registros Totales: {len(df)}")
    print(f"- Columnas Totales: {len(df.columns)}")
    print(f"- Valores Nulos (NaN): {nulos_tecnicos}")
    print(f"- Valores Sin Información ('-'): {nulos_semanticos}")
    print(f"- Completitud Real: {((total_celdas - nulos_semanticos) / total_celdas) * 100:.2f}%")
    print("Nota: El dataset no tiene nulos técnicos, pero usa '-' en diagnósticos secundarios.\n")

    # 2. VALIDACIÓN DE RANGOS: EDAD
    print(f"--- 2. CORRECTITUD: RANGOS DE EDAD ---")
    edad_min = df['Edad en años'].min()
    edad_max = df['Edad en años'].max()
    print(f"- Rango detectado: [{edad_min} - {edad_max}] años")

    # Criterio de Ingeniería: Edades negativas o sobre 110 años
    fuera_rango = df[(df['Edad en años'] < 0) | (df['Edad en años'] > 110)]
    print(f"- Registros fuera de rango lógico (0-110 años): {len(fuera_rango)}")
    if len(fuera_rango) > 0:
        print(f"  * Acción: Se identifica outlier de {edad_max} años para tratamiento en pre-procesamiento.")
    print("")

    # 3. VALIDACIÓN DE CATEGORÍAS: SEXO
    print(f"--- 3. CORRECTITUD: CATEGORÍAS DE SEXO ---")
    categorias_sexo = df['Sexo (Desc)'].unique()
    print(f"- Categorías encontradas: {categorias_sexo}")

    conteo_sexo = df['Sexo (Desc)'].value_counts()
    print("- Frecuencias:")
    for cat, val in conteo_sexo.items():
        print(f"  * {cat}: {val} registros")

    # Verificación de consistencia
    if len(categorias_sexo) <= 2:
        print("  * Validación: No se detectan errores de digitación en la variable Sexo.")
    else:
        print("  * Alerta: Se detectan más de 2 categorías, revisar posibles errores de ingreso.")

# Ejecutar el reporte
reporte_calidad_datos(df)
```
